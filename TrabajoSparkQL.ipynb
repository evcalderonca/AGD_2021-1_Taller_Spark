```python
# 2. Cree dos tablas SparkSQL y almacene el csv en su correspondiente tabla.
# Observación: tenga especial cuidado con los encabezados de los archivos CSV.
# Usted puede considerar cambiar los encabezados de los CSV originales
# Tu código a continuación...
#sentencia para crear las tablas
#ContextoSql.sql("""DROP TABLE EVALUACION """)
#ContextoSql.sql("""DROP TABLE ASISTENCIA """)

ContextoSql.sql("""
CREATE TABLE IF NOT EXISTS 
EVALUACION (
            MarcaTemporal VARCHAR(100), 
            NombreUsuario VARCHAR(100),
            EquipoEvaluar VARCHAR(100),
            Pregunta1 VARCHAR(10),
            Pregunta2 VARCHAR(10),
            Pregunta3 VARCHAR(10),
            Pregunta4 VARCHAR(10),
            Pregunta5 VARCHAR(10),
            Pregunta6 VARCHAR(10),
            Pregunta7 VARCHAR(10),
            Pregunta8 VARCHAR(10),
            Pregunta9 VARCHAR(10),
            Pregunta10 VARCHAR(10),
            Pregunta11 VARCHAR(10),
            Pregunta12 VARCHAR(10),
            Pregunta13 VARCHAR(10),
            Pregunta14 VARCHAR(10),
            Pregunta15 VARCHAR(10),
            Pregunta16 VARCHAR(10),
            Observaciones VARCHAR(100)
            ) 
USING CSV OPTIONS ( 
    header='true', 
    nullvalue='NA', 
    timestampFormat=\"yyyy-MM-dd'T'HH:mm:ss\", 
    path='/tmp/EvaluacionPitch.csv')
""");

ContextoSql.sql("""
CREATE TABLE IF NOT EXISTS 
ASISTENCIA (
            MarcaTemporal VARCHAR(100), 
            NombreUsuario VARCHAR(100),
            EquipoPertenece VARCHAR(100),
            EquipoExponer VARCHAR(100)
            ) 
USING CSV OPTIONS ( 
    header='true', 
    nullvalue='NA', 
    path='/tmp/AsistenciaPitch.csv')
""");
```

```python
# 2. Consulte el listado total de estudiantes (correos electrónicos) del 
# curso de Ingeniería de Software, ordenados alfabéticamente
# Tu código a continuación...
#consultas SQL
ContextoSql.sql("SELECT DISTINCT REPLACE(NombreUsuario,' ','') AS NombreUsuario from EVALUACION ORDER BY 1").show()
```

```python
# 3. Consulte la cantidad de asistencias registradas por estudiante; además, la fecha y hora de la primera asistencia
# y la fecha y hora de la última asistencia
# Tu código a continuación...
ContextoSql.sql("""SELECT REPLACE(NombreUsuario,' ','') AS NombreUsuario
                , COUNT(*) AS CantidadAsistencias
                , MIN(MarcaTemporal) AS FechaHoraPrimeraAsistencia
                , MAX(MarcaTemporal) AS FechaHoraUltimaAsistencia
            FROM ASISTENCIA 
            GROUP BY NombreUsuario
            ORDER BY 1 """).show()
```

```python
# 4. Consulte el listado de estudiantes que asistieron a 2 presentaciones o menos (una).
# Tu código a continuación...
ContextoSql.sql("SELECT NombreUsuario from ASISTENCIA GROUP BY NombreUsuario HAVING COUNT(NombreUsuario)<=2").show()
```

```python
# 5. Consulte el listado de estudiantes que no asistieron a ninguna presentación.
# Tu código a continuación...
ContextoSql.sql("""SELECT DISTINCT evaluacion.NombreUsuario 
                FROM EVALUACION evaluacion
                LEFT JOIN ASISTENCIA asistencia ON evaluacion.NombreUsuario = asistencia.NombreUsuario
                WHERE asistencia.NombreUsuario IS NULL""").show()
```

```python
# 6. Consulte los integrantes por cada equipo al que pertenecen.
# Tu código a continuación...
ContextoSql.sql("""SELECT DISTINCT NombreUsuario,REPLACE(EquipoPertenece,'Pertenezco a: ','') AS EquipoPertenece
                FROM ASISTENCIA
                ORDER BY EquipoPertenece,NombreUsuario""").show()
```

```python
# 7. Consulte la cantidad de asistentes por presentación, sin considerar los asistentes que pertenecen 
# al equipo que realizó la presentación.
# Tu código a continuación...
ContextoSql.sql("""SELECT REPLACE(asistencia.EquipoExponer,'Expone: ','') AS EquipoExponer,COUNT(asistencia.NombreUsuario) AS CantidadAsistentes
                FROM ASISTENCIA asistencia
                WHERE REPLACE(asistencia.EquipoPertenece,'Pertenezco a: ','') != REPLACE(asistencia.EquipoExponer,'Expone: ','') 
                GROUP BY asistencia.EquipoExponer
                ORDER BY asistencia.EquipoExponer""").show()
```

```python
# 8. Consutar cuáles integrantes evaluaron a su propio equipo. Estas evaluaciones no serán válidas, pues un
# integrante no puede evaluar a su propio equipo.
# Tu código a continuación...
ContextoSql.sql("""SELECT DISTINCT evaluacion.NombreUsuario, evaluacion.EquipoEvaluar
                FROM EVALUACION evaluacion
                INNER JOIN ASISTENCIA asistencia 
                    ON evaluacion.NombreUsuario = asistencia.NombreUsuario
                    AND REPLACE(asistencia.EquipoPertenece,'Pertenezco a: ','') = evaluacion.EquipoEvaluar""").show()
```
